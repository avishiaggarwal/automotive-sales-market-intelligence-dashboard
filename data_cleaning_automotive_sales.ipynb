```python
df.drop(columns=['vin'], inplace=True)
```

```python
df.dropna(subset=[
    'make',
    'model',
    'mmr',
    'sellingprice',
    'saledate'
], inplace=True)
```

```python
df['trim'] = df['trim'].fillna('Unknown')
df['body'] = df['body'].fillna('Unknown')
df['transmission'] = df['transmission'].fillna('Unknown')
df['color'] = df['color'].fillna('Unknown')
df['interior'] = df['interior'].fillna('Unknown')
```

```python
df['condition'] = df['condition'].fillna(df['condition'].median())
df['odometer'] = df['odometer'].fillna(df['odometer'].median())
```

```python
df.isnull().sum()
```

```python
df['saledate'].head(10)
```

```python
df['Sale_Year'] = df['saledate'].dt.year
df['Sale_Month'] = df['saledate'].dt.month_name()
df['Sale_Month_Num'] = df['saledate'].dt.month
df['Sale_Quarter'] = df['saledate'].dt.quarter
df['Sale_Day'] = df['saledate'].dt.day_name()
```

```python
df[['saledate','Sale_Year','Sale_Month','Sale_Quarter']].head()
```

```python
df['Vehicle_Age'] = df['Sale_Year'] - df['year']
df['Vehicle_Age'] = df['Vehicle_Age'].clip(lower=0)
```

```python
df[['year','Sale_Year','Vehicle_Age']].head()
```

```python
df['Mileage_Band'] = pd.cut(
    df['odometer'],
    bins=[0, 25000, 50000, 100000, 150000, float('inf')],
    labels=[
        'Very Low',
        'Low',
        'Moderate',
        'High',
        'Very High'
    ]
)
```

```python
df['Price_Range'] = pd.cut(
    df['sellingprice'],
    bins=[0,5000,10000,20000,40000,60000,float('inf')],
    labels=[
        'Entry',
        'Budget',
        'Economy',
        'Mid-Range',
        'Premium',
        'Luxury'
    ]
)
```

```python
def vehicle_condition(x):
    if x >= 4:
        return 'Excellent'
    elif x >= 3:
        return 'Good'
    elif x >= 2:
        return 'Fair'
    else:
        return 'Poor'

df['Vehicle_Condition'] = df['condition'].apply(vehicle_condition)
```

```python
df['Price_Difference'] = df['sellingprice'] - df['mmr']
```

```python
def market_status(x):
    if x > 1000:
        return 'Sold Above Market'
    elif x < -1000:
        return 'Sold Below Market'
    else:
        return 'Sold Near Market'

df['Market_Status'] = df['Price_Difference'].apply(market_status)
```

```python
def vehicle_condition(score):
    if score >= 40:
        return "Excellent"
    elif score >= 30:
        return "Good"
    elif score >= 20:
        return "Fair"
    else:
        return "Poor"

df['Vehicle_Condition'] = df['condition'].apply(vehicle_condition)
```
